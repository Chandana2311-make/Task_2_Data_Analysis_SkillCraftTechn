```python
import pandas as pd
```

```python
global_superstore = pd.read_csv("Global_Superstore2.csv", encoding="latin1")
```

```python
global_superstore.head()
```

```python
global_superstore.info()
```

```python
global_superstore.isnull().sum()
```

```python
global_superstore["Postal Code"] = global_superstore["Postal Code"].fillna(0)
```

```python
global_superstore.isnull().sum()
```

```python
global_superstore["Order Date"] = pd.to_datetime(global_superstore["Order Date"])
```

```python
global_superstore["Ship Date"] = pd.to_datetime(global_superstore["Ship Date"])
```

```python
global_superstore.dtypes
```

```python
global_superstore = global_superstore.drop_duplicates()
```

```python
global_superstore.shape
```

```python
global_superstore.to_csv("cleaned_global_superstore.csv", index=False)
```
